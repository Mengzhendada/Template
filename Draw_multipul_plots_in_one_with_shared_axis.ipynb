```python
import numpy as np
import matplotlib.pyplot as plt
x_axis=[.2,.3,.4,.5,.6,.7]
fig, axs = plt.subplots(2,2,figsize=(10,8))
fig.subplots_adjust(hspace=0)
fig.subplots_adjust(wspace=0)
axs[0][0].plot(x_axis,np.asarray(x_axis),'bo')
for i_x in x_axis:
    axs[0][0].plot([i_x,i_x],[i_x-0.1*i_x,i_x+0.1*i_x],color='blue',marker='_')
axs[0][0].set_ylabel('Y name',fontsize=16)
axs[0][0].set_ylim(.0002,.75)#Use 'bad number' for the edge to avoid overlap
axs[0][0].fill_between(x_axis,np.asarray(0.0002),np.asarray(0.0002)+0.1*np.asarray(x_axis),color='blue',hatch='\\\\\\\\',alpha=0.25,linewidth=0)
axs[0][0].tick_params(direction='in',labelbottom=False)
axs[0][0].text(0.5,0.9,r'plot {}'.format(1), horizontalalignment='center', verticalalignment='center', transform=axs[0][0].transAxes)
axs[0][1].plot(x_axis,np.asarray(x_axis)**2,'bo')
axs[0][1].set_ylim(.0002,.75)
axs[0][1].fill_between(x_axis,np.asarray(0.0002),np.asarray(0.0002)+0.1*np.asarray(x_axis)**2,color='blue',hatch='\\\\\\\\',alpha=0.25,linewidth=0)
axs[0][1].tick_params(direction='in',labelbottom=False,labelleft=False)
axs[0][1].text(0.5,0.9,r'plot {} for $y=x^2$'.format(2), horizontalalignment='center', verticalalignment='center', transform=axs[0][1].transAxes)
axs[1][0].plot(x_axis,np.asarray(x_axis)**3,'bo')
axs[1][0].set_ylim(.0002,.44)
axs[1][0].fill_between(x_axis,np.asarray(0.0002),np.asarray(0.0002)+0.1*np.asarray(x_axis)**3,color='blue',hatch='\\\\\\\\',alpha=0.25,linewidth=0)
axs[1][0].tick_params(direction='in')
axs[1][0].text(0.5,0.9,r'plot {} for $y=x^3$'.format(3), horizontalalignment='center', verticalalignment='center', transform=axs[1][0].transAxes)
axs[1][1].plot(x_axis,np.asarray(x_axis)**4,'bo')
axs[1][1].set_ylim(.0002,.44)
axs[1][1].set_xlabel('x name',fontsize=16)
axs[1][1].fill_between(x_axis,np.asarray(0.0002),np.asarray(0.0002)+0.1*np.asarray(x_axis)**4,color='blue',hatch='\\\\\\\\',alpha=0.25,linewidth=0)
axs[1][1].tick_params(direction='in',labelleft=False)
axs[1][1].text(0.5,0.9,r'plot {} for $y=x^4$'.format(4), horizontalalignment='center', verticalalignment='center', transform=axs[1][1].transAxes)
plt.savefig('Draw.pdf')
```
